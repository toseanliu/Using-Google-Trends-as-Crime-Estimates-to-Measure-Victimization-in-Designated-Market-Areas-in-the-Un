```python
import pandas as pd
import os
import numpy as np
import re
import pyreadstat
import matplotlib. pyplot as plt
from scipy import stats
```

```python
def combine_all_files(path, file_type = "csv",  on = "FIPS"):
    name_list = os.listdir(path) 
    if ".ipynb_checkpoints" in name_list:
        name_list.remove(".ipynb_checkpoints")
    if "desktop.ini" in name_list:
        name_list.remove("desktop.ini")
    data = []
    for i in name_list:
        if len(re.findall(f".*.{file_type}", i)) > 0:
            x = pd.read_csv(path + i, engine = "python").dropna(axis = 0, how = "all")
            data.append(x)
    data = pd.concat(data)
    data = data.reset_index().drop(columns = {"index"})
    return data
```

```python
def processing_ACS_data(df, year_name):
    df["FIPS"] = df["FIPS"].astype(int).astype(str)

    df["% Total Population: Male: 15 to 24 Years"] = \
    df["% Total Population: Male: 15 to 17 Years"] + \
    df["% Total Population: Male: 18 to 24 Years"]

    df = df.drop(["% Total Population: Male: 15 to 17 Years", 
                                        "% Total Population: Male: 18 to 24 Years", 
                                        "% Total Population: Male: 25 to 34 Years"], axis = 1)

    df["% Total: Moved From Different County Within Same State, Different State, Abroad"] = \
    df["% Total: Moved From Different County Within Same State"] + \
    df["% Total: Moved From Different State"] + \
    df["% Total: Moved From Abroad"]  

    df = df.drop(["% Total: Moved From Different County Within Same State", 
                                        "% Total: Moved From Different State", 
                                        "% Total: Moved From Abroad"], axis = 1)

    df_new = df.groupby("FIPS").mean().reset_index()

    df_new.FIPS = df_new.FIPS.astype(int).astype(str)

    df_new["Population Density (Per Sq. Mile)"] = np.log(df_new["Population Density (Per Sq. Mile)"])                           

    df_new["Mobility Index"] = df_new[["% Population 1 Year and Over in Households in the United States: Householder Lived in Renter-Occupied Housing Units: ", 
                                                         "% Total: Moved From Different County Within Same State, Different State, Abroad"]].dropna().apply(stats.zscore).sum(axis = 1)

    df_new["Economic Deficiency Index"] = df_new[["% Population 16 Years and Over: in Labor Force: Civilian: Unemployed", 
                                                 "% Families: Income Below Poverty Level",  
                                                   "% Households: Households with One or More People Under 18 Years: Family Households: Other Family (Single Parent)"]].dropna().apply(stats.zscore).sum(axis = 1)

    df_new["Single_Parent_and_Poverty"] = df_new[["% Families: Income Below Poverty Level", 
                                                                     "% Households: Households with One or More People Under 18 Years: Family Households: Other Family (Single Parent)"
                                                                    ]].dropna().apply(stats.zscore).sum(axis = 1)

    df_new["Immigration Index"] = df_new[["% Foreign-Born Population", "% Total Population: Hispanic or Latino"]].dropna().apply(stats.zscore).sum(axis = 1)

    df_new["Heterogeneity Index"] = (df_new["% Total Population: White Alone"]**2+\
    df_new["% Total Population: Hispanic or Latino"]**2+\
    df_new["% Total Population: American Indian and Alaska Native Alone"]**2+\
    df_new["% Total Population: Asian Alone"]**2+\
    df_new["% Total Population: Native Hawaiian and Other Pacific Islander Alone"]**2+\
    df_new["% Total Population: Black or African American Alone"]**2)/10000

    df_new["Heterogeneity Index"] = 1 - df_new["Heterogeneity Index"]

    df_new["Heterogeneity Index2"] = \
    (df_new["Heterogeneity Index"]-df_new["Heterogeneity Index"].mean())/df_new["Heterogeneity Index"].std() + \
    (df_new["% Foreign-Born Population"]-df_new["% Foreign-Born Population"].mean())/df_new["% Foreign-Born Population"].std()

    df_new[["% Families: Income Below Poverty Level"]].dropna().apply(stats.zscore)
    df_new.columns = df_new.columns+year_name
    df_new = df_new.rename(columns = {("FIPS"+year_name):"FIPS"})

    return df_new
```

```python
def processing_health_data(df, year_name):
    df = df[["FIPS",
                                               "Percent Drinking Adults (Persons 18 Years and Over)",
                                              "Percent Current Smokers (Persons 18 Years and Over)",
                                              "Drug Poisoning Mortality Rate per 100,000 Population",
                                               "Child Mortality Rate per 1,000 Population",
                                               "Premature Age-adjusted Mortality Rate per 100,000 Population"]].\
                                                apply(pd.to_numeric, errors='coerce').replace([np.inf, -np.inf], np.nan).dropna()

    df.FIPS = df.FIPS.astype(int).astype(str)

    df_new = df.groupby("FIPS").mean()

    df_new = df_new.reset_index()

    df_new["FIPS"] = df_new["FIPS"].astype(int).astype(str)

    df_new

    df_new["Health Index"] = df_new[[  "Child Mortality Rate per 1,000 Population",
                                                   "Premature Age-adjusted Mortality Rate per 100,000 Population"]].dropna().apply(stats.zscore).sum(axis = 1)
    df_new = df_new.drop([ "Child Mortality Rate per 1,000 Population",
                                                   "Premature Age-adjusted Mortality Rate per 100,000 Population",
                                                  ], axis = 1)
    df_new.columns = df_new.columns+year_name
    
    df_new = df_new.rename(columns = {("FIPS"+year_name):"FIPS"})

    return df_new
    
```

```python
def proessing_act_ucr_data(df, year_name):
    df["FIPS"] = df["fips_state_county_code"]

    df["FIPS"] = pd.to_numeric(df["FIPS"], errors='coerce')

    df_new = df.groupby("FIPS").mean().reset_index()

    act_crime = df_new[["FIPS", "actual_mtr_veh_theft_total", "actual_theft_total", "actual_rape_total"]].dropna()

    act_crime.FIPS = act_crime.FIPS.astype(int).astype(str)
    
    if year_name == "_04_18":
        df_new_1 = act_crime.merge(countypop_04_18[["FIPS","county_pop_04_18"]], how = "inner", on = "FIPS").rename(
        columns = {"county_pop_04_18":"Total Population"})
    elif year_name == "_04_08":
        df_new_1 = act_crime.merge(countypop_04_18[["FIPS","county_pop_04_08"]], how = "inner", on = "FIPS").rename(
        columns = {"county_pop_04_08":"Total Population"})
    elif year_name == "_09_13":
        df_new_1 = act_crime.merge(countypop_04_18[["FIPS","county_pop_09_13"]], how = "inner", on = "FIPS").rename(
        columns = {"county_pop_09_13":"Total Population"})
    elif year_name == "_14_18":
        df_new_1 = act_crime.merge(countypop_04_18[["FIPS","county_pop_14_18"]], how = "inner", on = "FIPS").rename(
        columns = {"county_pop_14_18":"Total Population"})
    elif year_name == "_04_11":
        df_new_1 = act_crime.merge(countypop_04_18[["FIPS","county_pop_04_11"]], how = "inner", on = "FIPS").rename(
        columns = {"county_pop_04_11":"Total Population"})
    elif year_name == "_12_18":
        df_new_1 = act_crime.merge(countypop_04_18[["FIPS","county_pop_12_18"]], how = "inner", on = "FIPS").rename(
        columns = {"county_pop_12_18":"Total Population"})


    df_new_1.index = df_new_1.FIPS

    df_new_1 = df_new_1.drop("FIPS", axis = 1)

    df_new_rate = (df_new_1.iloc[:,:].div(df_new_1["Total Population"], axis=0))*100000

    df_new_rate = df_new_rate.reset_index().drop("Total Population", axis = 1)
    
    df_new_rate.columns = df_new_rate.columns+year_name
    
    df_new_rate = df_new_rate.rename(columns = {("FIPS"+year_name):"FIPS"})

    return df_new_rate
    
```
